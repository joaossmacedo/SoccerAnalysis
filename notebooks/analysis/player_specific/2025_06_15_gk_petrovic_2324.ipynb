## Getting data

```python
df_shots = pd.read_csv(path_shots)
df_shots.head(2)
```

```python
df_player_matches = pd.read_csv(path_player_matches)
df_player_matches.head(2)
```

## Filtering Petrovic

```python
df_petrovic_games = df_player_matches.copy(True)
```

```python
df_petrovic_games = df_petrovic_games[df_petrovic_games['player'] == 'Đorđe Petrović']
df_petrovic_games = df_petrovic_games[df_petrovic_games['season'] == 2324]
df_petrovic_games = df_petrovic_games[df_petrovic_games['min'] >= 90]
len(df_petrovic_games)
```

```python
df_petrovic_games.head(1)
```

## Filtering Petrovic shots faced

```python
df_petrovic_shots = df_shots.merge(
  df_petrovic_games[['game']], on=['game'], how='inner'
)
df_petrovic_shots = df_petrovic_shots[df_petrovic_shots['team'] != 'Chelsea']
print(len(df_petrovic_shots))
df_petrovic_shots.head(1)
```

## Grouping data

```python
df_petrovic_shots['goal'] = (df_petrovic_shots['outcome'] == 'Goal').astype('int')
df_petrovic_shots['PSxG'] = df_petrovic_shots['PSxG'].fillna(0)
df_petrovic_shots.head(1)
```

```python
PSxG = sum(df_petrovic_shots['PSxG'])
goals = sum(df_petrovic_shots['goal'])
ratio = (goals / PSxG) - 1
print(f'PSxG: {PSxG}')
print(f'goals: {goals}')
print(f'ratio: {ratio:.2%}')
```

```python
df_petrovic_shots_non_deflected = df_petrovic_shots[df_petrovic_shots['is_deflected'] == 0]

PSxG = sum(df_petrovic_shots_non_deflected['PSxG'])
goals = sum(df_petrovic_shots_non_deflected['goal'])
ratio = (goals / PSxG) - 1
print(f'non deflected PSxG: {PSxG}')
print(f'non deflected goals: {goals}')
print(f'non deflected ratio: {ratio:.2%}')
```

```python
xG = sum(df_petrovic_shots['xG'])
goals = sum(df_petrovic_shots['goal'])
ratio = (goals / xG) - 1
print(f'xG: {xG}')
print(f'goals: {goals}')
print(f'ratio: {ratio:.2%}')
```
